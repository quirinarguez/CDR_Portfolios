```python
# Extract range into dictionary with lists: dict = 'LandL':[1,0.25]
def extract_range(all_metrics_ranges,metric_interest,minimum,maximum):
    '''
    Extract a subset from a range based on some criteria, namely 'minimum' and 'maximum'. 
    A few examples for a metric whose range is [1, 10]:
    -  if minimum = 1 and maximum = 0.25 --> range of interest is the first quarter of the distribution, i.e. [1, 2.5]
    -  if minimum = 0.25 and maximum = 1 --> range of interest is the last quarter of the distribution, i.e. [7.5, 10]
    -  if minimum = 0.5 and maximum = 1 --> range of interest is the last half of the distribution, i.e. [5, 10]
  
    Parameters:
    - all_metrics_ranges: dictionary of all available metrics in the form of: 'metric':[[min, max]] 
    - metric_interest: from the dictionary above, which metric is of interest.
    - minimum & maximum: criteria to select range, as per above. 
    '''

    minimum_value = all_metrics_ranges[metric_interest][0][0]
    maximum_value = all_metrics_ranges[metric_interest][0][1]
    if minimum == 1:
        interest = [minimum_value, minimum_value+(maximum_value-minimum_value)*maximum]
    elif maximum == 1:
        interest = [minimum_value+(maximum_value-minimum_value)*minimum, maximum_value]
    else: 
        interest = [minimum_value+(maximum_value-minimum_value)*minimum, minimum_value+(maximum_value-minimum_value)*maximum]
    
    return {metric_interest : [interest]}
```

```python
# Defining a function to plot the range of interest
def plot_horizontal_box(parameter, whole_range, range_interest, unit, divi=1, ax=None, box_height=0.5, color='blue'):
    """
    Plot a horizontal box with a colored bar inside representing a subset of the range.
    
    Parameters:
    - parameter: metric of interest
    - whole_range[parameter]: List of [A, B] values, the full range.
    - range_interest[parameter]: List of [C, D] values, the subset of the range.
    - unit[parameter]: units to be plotted under the bar
    - divi[parameter]: scaling factor
    - ax: axis where to plot. If 'None', function will create its own axis.
    - box_height: Height of the box (default 0.5).
    - color: Color of the subset range bar (default 'blue').
    """

    whole_range = whole_range[parameter]
    range_interest = range_interest[parameter]
    unit = unit[parameter]
    divi = divi[parameter]

    whole_range = [[x/divi for x in whole_range[0]]]
    range_interest = [[x/divi for x in range_interest[0]]]

    if ax is None:
        ax = plt.gca()
        plt.tight_layout()

    for i, ((A, B), (C, D)) in enumerate(zip(whole_range, range_interest)):
        # Draw the black box outline for the full range
        rect = patches.Rectangle((A, i - box_height / 2), B - A, box_height, 
                                  edgecolor='black', facecolor=(0,0,0,0), lw=3, zorder=2)
        ax.add_patch(rect)
        
        # Draw the colored bar inside the box representing the range of interest
        rect_colored = patches.Rectangle((C, i - box_height / 2), D - C, box_height, 
                                         facecolor=color, lw=0, zorder=1)
        ax.add_patch(rect_colored)
    
    # Adjust plot appearance
    ax.set_yticks([])
    ax.set_xlabel("[%s]" %unit)
    mini = min(min(A for A, B in whole_range), min(C for C, D in range_interest))
    maxi = max(max(B for A, B in whole_range), max(D for C, D in range_interest))
    ax.set_xlim(mini, maxi)
    x_ticks = np.linspace(mini,maxi, 5)

    ax.set_ylim(-0.25, len(whole_range) - 0.5)
    ax.grid(False)
    ax.spines[['top','right','left','bottom']].set_visible(False)
    #ax.spines['bottom'].set_color('black')

    # Setting x-axis ticks: 
    if maxi <= 1: ax.set_xticks(ticks=x_ticks, labels=[round(_, 2) for _ in x_ticks])
    ax.set_xticks(ticks=x_ticks, labels=[round(_) for _ in x_ticks])
    
    return ax
```

## Simplified Portfolios

### Defining Bins
Infeasible futures ("NaN") have been dropped.

Bins are created using results by 2100.

```python
data = removed_per_CDR.copy()#.div(1e9)
data = data.dropna(axis='index')

# Set the number of bins
n_bins = 4

# Create a dataframe that will store bins
bins = pd.DataFrame(index=i, columns = list(range(0,n_bins+1)))

for CDR in data.columns:
    values, bined = pd.cut(data[CDR], bins=n_bins, retbins=True)
    bins.loc[CDR] = bined

    '''pd.cut: The range of data[CDR] is extended by .1% on each side to include 
    the minimum and maximum values of data[CDR]. The minimum is therefore too small: we replace
    bin 0 with the minimum from the data +1 to allow for numerical variations, as this is 
    simply an upper bound'''

    bins.loc[CDR,0] = data.min()[CDR]+1

bins.style.format(precision=0, thousands=" ", decimal=".")
```

### Creating table with simplified portfolios for 1 time period

```python
# Create a table with max capacity per CDR option per future for 3 different time periods

simplified = pd.DataFrame()
years = [2050, 2100]

for futureID in data.index:
    results_in_f = pd.DataFrame()
    results_in_f.index = pd.MultiIndex.from_tuples([(futureID, i[n]) for n in range(0, len(i))],names=['future_id','CDR_opt'])
    
    for year in years:
        if year == 2050: data = removed_per_CDR_2050
        elif year == 2100: data = removed_per_CDR
        
        for CDR in i:
            # Check for Nan
            if np.isnan(data.loc[futureID,CDR]) == True:
                results_in_f.loc[(futureID,CDR),year] = np.nan
            else:
                bin_n = bins.columns.min() # Start search with smallest bin
                found = False 
                while found == False:
                    if data.loc[futureID,CDR] <= bins.loc[CDR, bin_n]: # If the value is smaller or equal than the bin value, search finishes
                        found = True
                        results_in_f.loc[(futureID, CDR),year] = bins.loc[CDR, bin_n]
                    else:
                        bin_n += 1

    simplified = pd.concat([simplified,results_in_f])

simplified = simplified.dropna(axis=0, how="all")
simplified.head(15).style.format(precision=0, thousands=" ", decimal=".")
```

### Identifying repeated portfolios

Here the output dataframe "simplified_names" has one row per future, rather than a multi-index as for the "simplified" dataframe. Each case includes a list with the portfolio description as a string. This way, duplicates can easily be identified.

```python
simplified_names = pd.DataFrame()

for futureID, CDR in simplified.groupby(level=0):
    results_in = pd.DataFrame()
    for n in years: # years is defined in the previous cell
        results_in.loc[futureID,n] = str(list(simplified.loc[(futureID),n].to_dict().items()))       
    
    # Add column with cumulative removals per future ID 
    if n == 2100: results_in.loc[futureID,'GtCO2'] = all_imported_data.loc[futureID, 'Removals']/(1e9)
    simplified_names = pd.concat([simplified_names,results_in])
    
simplified_names
# The index of 'simplified_names' is the future_ID
```

```python
# Creating a dataframe that counts how many times a portfolio is repeated:
simple_portfolios_2050 = pd.DataFrame(simplified_names[2050].value_counts())
simple_portfolios_2100 = pd.DataFrame(simplified_names[2100].value_counts())

year = 2050
# For each time period:
for time in [simple_portfolios_2050, simple_portfolios_2100]:
    if time is simple_portfolios_2100: year = 2100

    # For each portfolio (row) in "simple_portfolios_":
    for portfolio in time.index:
        repeated = simplified_names[year].isin([portfolio])
        # Find the index of the first occurrence of 'portfolio' in the DataFrame "simplified_names"
        time.loc[portfolio,"Example_FutureID"] = repeated.idxmax() # 'idxmax' Return index of first occurrence of maximum over requested axis.
                
        repeated = list(repeated[repeated].index)  # List on futureIDs that have the occurrence 'portfolio'

        # Find the range removals delivered across all futures that have the occurrence 'portfolio'
        maxi = 400
        mini = 1100
        for futureID in repeated:
            removed = total_future.loc[futureID,"Removals"]/1e9
            if removed <= mini:
                mini = removed
            if removed >= maxi:
                maxi = removed
        
        time.loc[portfolio,"Removals_Min"] = mini 
        time.loc[portfolio,"Removals_Max"] = maxi

simple_portfolios_2100.head(8)
```

## Histograms for Scenarios

```python
# Defining a function to plot the range of interest specifically for histograms
def plot_narrow_box(parameter, whole_range, range_interest, unit, divi=1, ax=None, box_height=0.5, color='blue'):

    whole_range = whole_range[parameter]
    range_interest = range_interest[parameter]
    unit = unit[parameter]
    divi = divi[parameter]

    whole_range = [[x/divi for x in whole_range[0]]]
    range_interest = [[x/divi for x in range_interest[0]]]

    """
    Plot a horizontal box with a colored bar inside representing a subset of the range.
    
    Parameters:
    - whole_range: List of [A, B] values, the full range.
    - range_interest: List of [C, D] values, the subset of the range.
    - box_height: Height of the box (default 0.5).
    - color: Color of the subset range bar (default 'blue').
    """
    #fig, ax = plt.subplots(figsize=(8, 2))
    if ax is None:
        ax = plt.gca()
        plt.tight_layout()

    for i, ((A, B), (C, D)) in enumerate(zip(whole_range, range_interest)):
        # Draw the black box outline for the full range
        rect = patches.Rectangle((A, i - box_height / 2), B - A, box_height, 
                                  edgecolor='black', facecolor=(0,0,0,0), lw=3, zorder=2)
        ax.add_patch(rect)
        
        # Draw the colored bar inside the box representing the range of interest
        rect_colored = patches.Rectangle((C, i - box_height / 2), D - C, box_height, 
                                         facecolor=color, lw=0, zorder=1)
        ax.add_patch(rect_colored)
    
    # Adjust plot appearance
    x_pos = (0, 0.27)
    ax.set_yticks([])
    ax.set_xlabel("[%s]" %unit)
    ax.xaxis.set_label_coords(0.5,0.15)

    mini = min(min(A for A, B in whole_range), min(C for C, D in range_interest))
    maxi = max(max(B for A, B in whole_range), max(D for C, D in range_interest))
    ax.set_xlim(mini, maxi)
    x_ticks = np.linspace(mini,maxi, 2)
    if mini >= 1000: x_ticks = np.linspace(mini,maxi, 2)

    ax.set_ylim(-0.25, len(whole_range) - 0.5)
    ax.grid(False)
    ax.spines[['top','right','left','bottom']].set_visible(False)
    #ax.spines['bottom'].set_color('black')

    # Setting x-axis ticks: 
    if maxi <= 1: ax.set_xticks(ticks=x_ticks, labels=[round(_, 2) for _ in x_ticks])
    ax.set_xticks(ticks=x_ticks, labels=[round(_) for _ in x_ticks], position = x_pos)
    ax.tick_params(bottom = False) 

    return ax
```
